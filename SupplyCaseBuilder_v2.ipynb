```python

# Import necessary libraries
import pandas as pd
import numpy as np
import tkinter as tk
from tkinter import filedialog, simpledialog
import os
import datetime
import calendar
import matplotlib.pyplot as plt


# Step 1: File Selection and Renaming
#   Function select_and_rename_files:
#   Opens a file dialog to select multiple .xlsx files.
#   Renames each selected file based on the scenario value from the "Info" sheet and the year from the "Time Series Table" sheet.

def select_and_rename_files():
    # Setup the root Tkinter window and hide it
    root = tk.Tk()
    root.withdraw()

    # Open a file dialog to select multiple .xlsx files
    file_paths = filedialog.askopenfilenames(title="Select Excel files", filetypes=[("Excel files", "*.xlsx")])

    for file_path in file_paths:
        # Load the scenario from the first tab
        df_scenario = pd.read_excel(file_path, sheet_name = "Info")  # Assumes scenario is in the first tab
        scenario_value = df_scenario.loc[3,1]  # Adjust column names as needed

        # Load the year from the second tab named "time series table"
        df_time_series = pd.read_excel(file_path, sheet_name="Time Series Table")
        year_value = df_time_series['Year'].iloc[0]  # Adjust based on actual structure

        # Create a new file name based on the scenario and year
        directory, old_name = os.path.split(file_path)
        name, ext = os.path.splitext(old_name)
        new_name = f"{name}_{scenario_value}_{year_value}{ext}"
        new_file_path = os.path.join(directory, new_name)

        # Rename the file
        os.rename(file_path, new_file_path)
        print(f"Renamed '{file_path}' to '{new_file_path}'")

# Step 2: Load Default Sheet Names
#   Function load_default_sheets: Loads default sheet names from a CSV file.
#   Define the path for the mapping file
mapping_file_path = 'last_file_paths.csv'
def load_default_sheets():
    if os.path.exists(mapping_file_path):
        return pd.read_csv(mapping_file_path, index_col='file_path').to_dict()['sheet_name']
    return {}

#   Function save_default_sheet: Saves the sheet name for a given file path to a CSV file.
def save_default_sheet(file_path, sheet_name):
    default_sheets = load_default_sheets()
    default_sheets[file_path] = sheet_name
    df = pd.DataFrame(list(default_sheets.items()), columns=['file_path', 'sheet_name'])
    df.to_csv(mapping_file_path, index=False)

# Function get_sheet_name: Retrieves the sheet name for a file from the user if it's not in the default sheet names.
def get_sheet_name(file_path, root):
    default_sheets = load_default_sheets()
    if file_path in default_sheets:
        return default_sheets[file_path]
    else:
        # Ask user for a new sheet name
        sheet_name = simpledialog.askstring("Sheet Selection", f"Enter the sheet name for {file_path.split('/')[-1]}:", parent=root)
        if sheet_name:
            save_default_sheet(file_path, sheet_name)
        return sheet_name

# Step 3: Load Data
# Function load_data:
#   Opens a file dialog to select multiple files.
#   Loads the specified sheet from each file into a DataFrame and concatenates them into one DataFrame.

def load_data():
    # Set up the root tkinter widget
    root = tk.Tk()
    root.withdraw()  # Hide the main window

    # Open file dialog and allow multiple file selection
    file_paths = filedialog.askopenfilenames(title="Select files", filetypes=(("Excel files", "*.xlsx"), ("All files", "*.*")))

    # Load and concatenate the data from selected files
    data_frames = []
    for file_path in file_paths:
        if file_path.endswith('.xlsx'):
            sheet_name = get_sheet_name(file_path, root)
            if sheet_name:
                df = pd.read_excel(file_path, sheet_name=sheet_name)
                data_frames.append(df)

    # Concatenate all DataFrames into one DataFrame
    if data_frames:
        combined_data = pd.concat(data_frames, ignore_index=True)
        print("Data successfully loaded and combined.")
        return combined_data
    else:
        print("No files selected or no valid sheets specified.")
        return None


# Step 4: Generate Monthly Columns
# Function format_month_year: Formats a year and month into a string "MMM-YYYY".
def format_month_year(year, month):
    """Generate a string in the format 'MMM-YYYY' from a year and month."""
    date = datetime.datetime(year, month, 1)  # The day is arbitrary; we just need the month and year.
    return date.strftime("%b-%Y")

# Function generate_monthly_columns: Generates a list of monthly column names in the format "MMM-YYYY" for a given range of years.
def generate_monthly_columns(start_year, end_year):
    """ Generate a list of strings representing each month of each year in the range from start_year to end_year. """
    monthly_cols = []
    for year in range(start_year, end_year + 1):
        for month in range(1, 13):  # Months from January (1) to December (12)
            monthly_col = f"{year}-{month:02d}"  # Format: YYYY-MM
            monthly_col = format_month_year(year, month)
            monthly_cols.append(monthly_col)

    return monthly_cols

# Step 5: Distribute Annual Data to Monthly Data

# Function distribute_annual_to_monthly_crosstab: Distributes annual production data evenly across monthly columns.
def distribute_annual_to_monthly_crosstab(df, start_year=2011, end_year=2060):
    # Iterate over each year, create monthly columns
    # Create an empty DataFrame to store results
    result_df = df[['GpuName', 'Scenario']].copy()

    for year in range(start_year, end_year + 1):
        annual_col = year
        days_in_month = [calendar.monthrange(year, month)[1] for month in range(1, 13)]
        total_days_year = sum(days_in_month)
        average_month_length = total_days_year / 12
        #monthly_cols = [f"{year}-{month:02d}" for month in range(1, 13)]
        monthly_cols = [format_month_year(year, month) for month in range(1, 13)]
       
        # Assuming annual production is evenly distributed across months
        for i, month_col in enumerate(monthly_cols):
            
            #Method 1
            # Adjust by dividing the annual production by total days and multiplying by the days in the month
            #result_df[month_col] = df[annual_col] / average_month_length * days_in_month[i]
            #Method 2
            # No adjustmnet by just using the same rate for the monthly.  
            result_df[month_col] = df[annual_col] 


    return result_df

# Step 6: Apply Central Moving Average
# Function apply_central_moving_average: Applies a central moving average to smooth monthly production data.
def apply_central_moving_average(df, monthly_cols, window_size,  edge_handling='pad'):
    """Applies central moving average with edge handling to a DataFrame across specified columns."""
    if window_size % 2 == 0:
        raise ValueError("Window size must be odd for central moving average")

    kernel = np.ones(window_size) / window_size
    pad_width = window_size // 2  # How much to pad on each side

    # Apply CMA for each row
    for index, row in df.iterrows():
        original_values = row[monthly_cols].fillna(0).to_numpy()  # Handle possible NaNs by assuming 0 production

        # Choose padding strategy
        if edge_handling == 'pad':
            padded_values = np.pad(original_values, pad_width, mode='edge')
        elif edge_handling == 'reflect':
            padded_values = np.pad(original_values, pad_width, mode='reflect')
        elif edge_handling == 'wrap':
            padded_values = np.pad(original_values, pad_width, mode='wrap')
        else:
            raise ValueError("Unsupported edge handling method")

        smoothed_values = np.convolve(padded_values, kernel, mode='valid')
        df.loc[index, monthly_cols] = smoothed_values

    return df

# Step 7: Check Annual Sum
# Function check_annual_sum: Compares the summed smoothed monthly production data to the original annual values to ensure consistency.
def check_annual_sum(df_annual, df_smoothed_monthly, start_year=2011, end_year=2060):
    """
    Compares the summed smoothed monthly production data (accounting for daily rates) back to the original annual values.
    Args:
    df_annual: DataFrame containing the original annual production data with integer year columns.
    df_smoothed_monthly: DataFrame containing the smoothed monthly production data.
    start_year: The start year of the data.
    end_year: The end year of the data.
    Returns:
    A DataFrame with the original annual, summed smoothed monthly, and their ratios (NaN where annual is zero or null).
    """
    results = pd.DataFrame(index=df_smoothed_monthly.index)
    
    for year in range(start_year, end_year + 1):
        monthly_cols = [format_month_year(year, month) for month in range(1, 13)]
        days_in_month = [calendar.monthrange(year, month)[1] for month in range(1, 13)]  # Number of days in each month
        days_in_year = sum(days_in_month)  # Total days in the year
        
        # Sum monthly data taking into account the number of days in each month
        for i, month_col in enumerate(monthly_cols):
            df_smoothed_monthly[month_col] *= days_in_month[i]
        
        # Sum monthly data taking into account the number of days in each month
        summed_monthly = df_smoothed_monthly[monthly_cols].sum(axis=1)
        
        # Calculate the ratio of summed monthly to original annual
        #ratio_col_name = f"{format_month_year(year, 6)}_ratio"  # 'MMM-YYYY_ratio'
        ratio_col_name = f"{year}_ratio" # 'YYYY_ratio'
        # Check if annual production is zero or null
        results[ratio_col_name] =  (df_annual[year]* days_in_year) / summed_monthly

           # Replace infinite values with NaN
        results.replace([np.inf, -np.inf], np.nan, inplace=True)

    # Calculate min, max, and average ratios for each GpuName
    summary_results = pd.DataFrame(index=results.index)
    summary_results['GpuName'] = df_smoothed_monthly['GpuName']
    summary_results['Min_Ratio'] = results.min(axis=1)
    summary_results['Max_Ratio'] = results.max(axis=1)
    summary_results['Avg_Ratio'] = results.mean(axis=1)
    
    # Print the summary results DataFrame
    print(summary_results[['GpuName', 'Min_Ratio', 'Max_Ratio', 'Avg_Ratio']])
    
    return results, summary_results


# Step 8: Recalibrate Monthly Production
# Function recalibrate_monthly_production: Adjusts monthly production rates to align with the original annual totals based on given ratios.
def recalibrate_monthly_production1(df_monthly, monthly_cols, ratio_matrix):
    """
    Adjusts monthly production rates using annual total comparison ratios, weighted by the number of days in each month relative to the average month length.

    Args:
    df_monthly (DataFrame): DataFrame containing smoothed monthly production data.
    monthly_cols (list): List of column names for the monthly production data in the format 'mmm-yyyy'.
    ratio_matrix (DataFrame): DataFrame containing ratios calculated from the annual sum checks, with years as integer columns.

    Returns:
    DataFrame: Adjusted DataFrame with recalibrated monthly production figures.
    """
    # Calculate the average month length and days per month dynamically
    year_days = {}
    for col in monthly_cols:
        year_part = int(col.split('-')[1])  # Extract the year part from 'mmm-yyyy'
        if year_part not in year_days:
            year_days[year_part] = sum(calendar.monthrange(year_part, month)[1] for month in range(1, 13))

    average_days_per_month = {year: days / 12 for year, days in year_days.items()}

    # Apply the ratio adjustments to the monthly columns
    for month_col in monthly_cols:
        year_part = int(month_col.split('-')[1])  # Year part as integer
        month_part = month_col.split('-')[0]  # Month part as string
        month_index = list(calendar.month_abbr).index(month_part)  # Convert month name to month number
        days_in_month = calendar.monthrange(year_part, month_index)[1]

        if  f"{year_part}_ratio" in ratio_matrix.columns:
            # Calculate adjusted ratio using the days in the month vs average
            adjusted_ratio = (days_in_month / average_days_per_month[year_part])
            print(adjusted_ratio)
            if month_col in df_monthly.columns:
                df_monthly[month_col] *= adjusted_ratio * (ratio_matrix.loc[0, f"{year_part}_ratio"] if isinstance(ratio_matrix, pd.DataFrame) else ratio_matrix[f"{year_part}_ratio"])

    return df_monthly


def recalibrate_monthly_production2(df_monthly, monthly_cols, ratio_matrix):
    """
    Adjusts monthly production rates to align more closely with the original annual totals based on given ratios.
    
    Args:
    df_monthly (DataFrame): DataFrame containing monthly production data.
    monthly_cols (list): List of column names for the monthly production data in the format 'mmm-yyyy'.
    ratio_matrix (DataFrame): DataFrame containing ratios with columns named as 'yyyy_ratio'.

    Returns:
    DataFrame: DataFrame with adjusted monthly production figures.
    """
    # Calculate average month length and days per month
    year_days = {}
    for col in monthly_cols:
        year = int(col.split('-')[1])
        if year not in year_days:
            year_days[year] = sum(calendar.monthrange(year, m)[1] for m in range(1, 13))
    average_days_per_month = {year: days / 12 for year, days in year_days.items()}

    # Apply the ratio adjustments to the monthly columns
    for month_col in monthly_cols:
        year = int(month_col.split('-')[1])  # Year part as integer
        month_name = month_col.split('-')[0]  # Month part as string
        if month_name not in calendar.month_abbr:
            print(f"Error: '{month_name}' is not a valid month abbreviation.")
            continue
        month_index = calendar.month_abbr.index(month_name)  # Convert month name to month number

        if month_index == 0:
            print(f"Month index error with month '{month_name}' in column '{month_col}'.")
            continue

        days_in_month = calendar.monthrange(year, month_index)[1]
        ratio_column_name = f"{year}_ratio"

        if ratio_column_name in ratio_matrix.columns:
            # Calculate adjusted ratio using the days in the month vs average
            adjusted_ratio = (days_in_month / average_days_per_month[year]) * ratio_matrix[ratio_column_name].iloc[0]
            if month_col in df_monthly.columns:
                df_monthly[month_col] *= adjusted_ratio
            print(f"Adjusted ratio for {month_col}: {adjusted_ratio}")
        else:
            print(f"Ratio column {ratio_column_name} not found in ratio_matrix")

    return df_monthly

def recalibrate_monthly_production(df_monthly, monthly_cols, ratio_matrix):
    """
    Adjusts monthly production rates to align more closely with the original annual totals based on given ratios.
    
    Args:
    df_monthly (DataFrame): DataFrame containing monthly production data.
    monthly_cols (list): List of column names for the monthly production data in the format 'mmm-yyyy'.
    ratio_matrix (DataFrame): DataFrame containing ratios with columns named as 'yyyy_ratio'.

    Returns:
    DataFrame: DataFrame with adjusted monthly production figures.
    """
    df_result = df_monthly

    for month_col in monthly_cols:
        year = int(month_col.split('-')[1])  # Year part as integer
        ratio_column_name = f"{year}_ratio"

        if ratio_column_name in ratio_matrix.columns:
            adjusted_ratio =  ratio_matrix[ratio_column_name]
            adjusted_ratio = adjusted_ratio.where(pd.notna(adjusted_ratio), 1)  # Replace NaNs with 1
            #df_monthly[month_col] *= adjusted_ratio * (days_in_month / average_days_per_month[year])
        
        df_result[month_col] = df_monthly[month_col] * adjusted_ratio * 0.033


    return df_result

# Step 9: Convert Units
#  Function convert_mmcfd_to_mcmd: Converts volume from million cubic feet per day to million cubic meters per day.

def convert_mmcfd_to_mcmd(volume_mmcfd):
    """Convert volume from million cubic feet per day to million cubic meters per day."""
    return volume_mmcfd * 0.0283168

    
# Assume df_annual and df_smoothed_monthly are defined elsewhere in your code
# Here's how you would call this function:
# results = check_annual_sum(df_annual, df_smoothed_monthly)
# print(results.head())


# Step 10: Plot Production Data
# Function plot_production: Plots the production data for a specified GPU and scenario, either in monthly or annual format.


def plot_production(data, GpuName, scenario_col, time_format, title_postfix):
    """
    Plots the production data for Brazil by scenario.

    Args:
    data (DataFrame): The production data, expected to contain a 'Country', 'Scenario', and time columns.
    GpuName (str):  filter by, default is 'Brazil'.
    scenario_col (str): Column name for the scenario, default is 'Scenario'.
    time_format (str): Format of the time data in the DataFrame ('monthly' or 'annual').
    """
    # Filter data for the specified country
    filtered_data = data[data['GpuName'] == GpuName]
    
    # Set the index to the appropriate columns for plotting
    if time_format == 'monthly':
        # Assume monthly data columns are like 'Jan-2011', 'Feb-2011', etc.
        time_cols = [col for col in filtered_data.columns if isinstance(col, str) and '-' in col]  # simple filter for 'MMM-YYYY'
    elif time_format == 'annual':
        # Assume annual data columns are just years like 2011, 2012, etc.
        time_cols = [col for col in filtered_data.columns if isinstance(col, int) or (isinstance(col, str) and col.isdigit())]

    # Plotting setup
    plt.figure(figsize=(12, 6))
    
    # Different scenarios
    scenarios = filtered_data[scenario_col].unique()
    for scenario in scenarios:
        scenario_data = filtered_data[filtered_data[scenario_col] == scenario]
        production_totals = scenario_data[time_cols].sum()
        
        if time_format == 'monthly':
            # Convert 'MMM-YYYY' to datetime for proper sorting and plotting
            production_totals.index = pd.to_datetime(production_totals.index, format='%b-%Y')
            production_totals = production_totals.sort_index()
        
        plt.plot(production_totals.index, production_totals.values, label=scenario)
    
    plt.title(f'Production Data for {GpuName} {title_postfix}')
    plt.xlabel('Time')
    plt.ylabel('Production Rate')
    plt.legend(title='Scenario')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()  # Adjust plot to make room for rotated x-axis labels
    plt.show()

# Function to plot gas operating cost data
def plot_cost(data, GpuName, scenario_col, time_format, title_postfix):
    """
    Plots the gas operating cost data for a given GpuName by scenario.

    Args:
    data (DataFrame): The gas operating cost data, expected to contain a 'GpuName', 'Scenario', and time columns.
    GpuName (str): Filter by GpuName.
    scenario_col (str): Column name for the scenario.
    time_format (str): Format of the time data in the DataFrame ('monthly' or 'annual').
    """
    # Filter data for the specified GpuName
    filtered_data = data[data['GpuName'] == GpuName]
    
    # Set the index to the appropriate columns for plotting
    if time_format == 'monthly':
        # Assume monthly data columns are like 'Jan-2011', 'Feb-2011', etc.
        time_cols = [col for col in filtered_data.columns if isinstance(col, str) and '-' in col]  # simple filter for 'MMM-YYYY'
    elif time_format == 'annual':
        # Assume annual data columns are just years like 2011, 2012, etc.
        time_cols = [col for col in filtered_data.columns if isinstance(col, int) or (isinstance(col, str) and col.isdigit())]

    # Plotting setup
    plt.figure(figsize=(12, 6))
    
    # Different scenarios
    scenarios = filtered_data[scenario_col].unique()
    for scenario in scenarios:
        scenario_data = filtered_data[filtered_data[scenario_col] == scenario]
        cost_totals = scenario_data[time_cols].sum()
        
        if time_format == 'monthly':
            # Convert 'MMM-YYYY' to datetime for proper sorting and plotting
            cost_totals.index = pd.to_datetime(cost_totals.index, format='%b-%Y')
            cost_totals = cost_totals.sort_index()
        
        plt.plot(cost_totals.index, cost_totals.values, label=scenario)
    
    plt.title(f'Gas Operating Cost for {GpuName} {title_postfix}')
    plt.xlabel('Time')
    plt.ylabel('Gas Operating Cost')
    plt.legend(title='Scenario')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()  # Adjust plot to make room for rotated x-axis labels
    plt.show()

# Step 11: Construct Supply Curves
# Function construct_supply_curve: Constructs a supply curve by sorting and cumulatively summing the data.
def construct_supply_curve(gpu_data):
    # Sort and cumulative sum logic here
    return gpu_data.sort_values(by='cost').cumsum()

# Step 12: Export Data
# Function export_data: Exports the given DataFrame to a CSV file.
def export_data(data, file_name):
    data.to_csv(file_name)

# Main workflow
#   Loads and processes production and asset mapping data.
#   Renames columns and converts units.
#   Distributes annual data to monthly data and applies smoothing.
#   Checks and recalibrates the data.
#   Plots the production data.
# Exports the results to an Excel file.

def main():

    file_path = "C:/Users/NingLin/OneDrive - RBAC, Inc"
    map_file = "SPV asset mapping v1.43.xlsx"
    asset_map = pd.read_excel(os.path.join(file_path, map_file), sheet_name="3. Vantage Asset Map Final")
    print(asset_map.head()) 

    #Import file names (without scenarios yet)
    #productionrate (unit: mmcfd) for gas
    #opcost total (unit: MMUSD): including fixed and variable opcosts for the entire asset operations
    #totalprodratege: production total gas equivalent rate (unit: mmcfd) total production rate for all products (oil, gas, liquids) in gas equivalent

    prod_file = "S&PGlobal_VantageTimeSeriesTable_20240424_productionrate"
    cost_file = "S&PGlobal_Vantage_TimeSeriesTable_20240528_opcost"
    totalprod_file = "S&PGlobal_Vantage_TimeSeriesTable_20240528_totalprodratege"

    prod_data = []
    cost_data = []
    totalprod_data = []
    maxrows = 5000
    scenarios = ["low", "base", "high"]

    # Load production data
    for ss in scenarios:
        full_path = os.path.join(file_path, prod_file + "_" + ss + ".xlsx")
        try:
            df = pd.read_excel(full_path, sheet_name="Time Series Table", nrows=maxrows)
            df['Scenario'] = ss
            prod_data.append(df)
        except ValueError as e:
            print(f"Error in file {full_path}: {e}")
            continue # skip the file and continue

    # Load cost data
    for ss in scenarios:
        full_path = os.path.join(file_path, cost_file + "_" + ss + ".xlsx")
        try:
            df = pd.read_excel(full_path, sheet_name="Time Series Table", nrows=maxrows)
            df['Scenario'] = ss
            cost_data.append(df)
        except ValueError as e:
            print(f"Error in file {full_path}: {e}")
            continue # skip the file and continue

    # Load total production data
    for ss in scenarios:
        full_path = os.path.join(file_path, totalprod_file + "_" + ss + ".xlsx")
        try:
            df = pd.read_excel(full_path, sheet_name="Time Series Table", nrows=maxrows)
            df['Scenario'] = ss
            totalprod_data.append(df)
        except ValueError as e:
            print(f"Error in file {full_path}: {e}")
            continue # skip the file and continue

    # Combine data for each category
    if prod_data:
        production_data = pd.concat(prod_data, ignore_index=True)
        print(production_data.head())
    else:
        print("No production data was loaded due to errors.")

    totalprod_data = pd.concat(totalprod_data, ignore_index=True)
    cost_data = pd.concat(cost_data, ignore_index=True)

    # Ensure data types match for merging (if necessary)
    production_data['Asset Name'] = production_data['Asset Name'].astype(str)
    totalprod_data['Asset Name'] = totalprod_data['Asset Name'].astype(str)
    cost_data['Asset Name'] = cost_data['Asset Name'].astype(str)
    asset_map['Asset Name'] = asset_map['Asset Name'].astype(str)

    # Rename the 'Year' column to 'Variable' in all dataframes
    production_data.rename(columns={'Year': 'Variable'}, inplace=True)
    totalprod_data.rename(columns={'Year': 'Variable'}, inplace=True)
    cost_data.rename(columns={'Year': 'Variable'}, inplace=True)
    
    year_cols = [year for year in range(2011, 2061)]  # years are handled as integers

    # Continue with the identical process for weighted_gas_opcost as for production data
    production_data[year_cols] = production_data[year_cols].applymap(convert_mmcfd_to_mcmd)
    totalprod_data[year_cols] = totalprod_data[year_cols].applymap(convert_mmcfd_to_mcmd)

    # Merge production, total production, and cost data with asset mapping
    production_data_mapped = pd.merge(production_data, asset_map[['Asset Name', 'GpuName']], on='Asset Name', how='inner')
    totalprod_data_mapped = pd.merge(totalprod_data, asset_map[['Asset Name', 'GpuName']], on='Asset Name', how='inner')
    cost_data_mapped = pd.merge(cost_data, asset_map[['Asset Name', 'GpuName']], on='Asset Name', how='inner')

 
    # Calculate the gas operation cost at the asset level
    for scenario in scenarios:
        prod_scenario = production_data_mapped[production_data_mapped['Scenario'] == scenario]
        totalprod_scenario = totalprod_data_mapped[totalprod_data_mapped['Scenario'] == scenario]
        cost_scenario = cost_data_mapped[cost_data_mapped['Scenario'] == scenario]

        for year in year_cols:
            split_ratio = prod_scenario[year] / totalprod_scenario[year]
            gas_opcost = split_ratio * cost_scenario[year]/prod_scenario[year]
            cost_data_mapped.loc[cost_data_mapped['Scenario'] == scenario, year] = gas_opcost 


   # Weighted average gas operation cost at GPU level
    weighted_gas_opcost = cost_data_mapped[['GpuName', 'Scenario'] + year_cols].copy()
    for year in year_cols:
        weighted_gas_opcost[year] = (
            cost_data_mapped[year] * production_data_mapped[year]
        ).groupby(cost_data_mapped['GpuName']).transform('sum') / production_data_mapped[year].groupby(production_data_mapped['GpuName']).transform('sum')

    print(weighted_gas_opcost.head())


    
        # Move 'Scenario' and 'GpuName' columns to the front
    columns = ['Scenario', 'GpuName'] + [col for col in production_data_mapped.columns if col not in ['Scenario', 'GpuName']]
    production_data_mapped = production_data_mapped[columns]

    print(production_data_mapped)
    print("Columns available in DataFrame:", production_data_mapped.columns.tolist())

    # Verifying year columns
    print("Expected year columns:", year_cols)

    production_data_mapped_truncated = production_data_mapped[['Scenario', 'GpuName'] + year_cols]

    production_gpu_cross = production_data_mapped_truncated.groupby(['GpuName', 'Scenario']).sum().reset_index()

    GpuName = "Australia-West Offshore"

    plot_production(production_gpu_cross, GpuName, 'Scenario', 'annual', 'Original volume')

    # Generate monthly columns for a given range
    monthly_cols = generate_monthly_columns(2011, 2060)

    production_gpu_cross_monthly = distribute_annual_to_monthly_crosstab(production_gpu_cross)
    plot_production(production_gpu_cross_monthly, GpuName, 'Scenario', 'monthly', 'Distributed monthly volume')

    production_gpu_cross_monthly2 = apply_central_moving_average(production_gpu_cross_monthly, monthly_cols, window_size=11)
    plot_production(production_gpu_cross_monthly2, GpuName, 'Scenario', 'monthly', 'Central MV Distributed monthly volume')

    ratio_matrix, summary_results = check_annual_sum(production_gpu_cross, production_gpu_cross_monthly2)
    production_gpu_cross_monthly3 = recalibrate_monthly_production(production_gpu_cross_monthly2, monthly_cols, ratio_matrix)
    plot_production(production_gpu_cross_monthly3, GpuName, 'Scenario', 'monthly', 'Calibrated Central MV Distributed monthly volume')

    # Apply central moving average again after recalibration
    production_gpu_cross_monthly4 = apply_central_moving_average(production_gpu_cross_monthly3, monthly_cols, window_size=11)
    plot_production(production_gpu_cross_monthly4, GpuName, 'Scenario', 'monthly', 'Smoothed Calibrated Central MV Distributed monthly volume')

    ratio_matrix1, summary_results1 = check_annual_sum(production_gpu_cross, production_gpu_cross_monthly4)

    production_gpu = pd.melt(production_gpu_cross, id_vars=['GpuName', 'Scenario'], value_vars=year_cols, var_name='Period', value_name='Production rate (Mcm/d)')
    production_gpu_monthly = pd.melt(production_gpu_cross_monthly4, id_vars=['GpuName', 'Scenario'], value_vars=monthly_cols, var_name='Period', value_name='Yearly Production (Mcm/d)')

    print("Unpivoted DataFrame:\n", production_gpu)
    print("Unpivoted DataFrame:\n", production_gpu_monthly)

     # Process weighted_gas_opcost similarly
    weighted_gas_opcost_truncated = weighted_gas_opcost[['GpuName', 'Scenario'] + year_cols]

    weighted_gas_opcost_cross = weighted_gas_opcost_truncated.groupby(['GpuName', 'Scenario']).sum().reset_index()

    plot_cost(weighted_gas_opcost_cross, GpuName, 'Scenario', 'annual', 'Weighted Gas OpCost Annual')

    weighted_gas_opcost_cross_monthly = distribute_annual_to_monthly_crosstab(weighted_gas_opcost_cross)
    plot_cost(weighted_gas_opcost_cross_monthly, GpuName, 'Scenario', 'monthly', 'Distributed Weighted Gas OpCost Monthly')

    weighted_gas_opcost_cross_monthly2 = apply_central_moving_average(weighted_gas_opcost_cross_monthly, monthly_cols, window_size=11)
    plot_cost(weighted_gas_opcost_cross_monthly2, GpuName, 'Scenario', 'monthly', 'Smoothed Distributed Weighted Gas OpCost Monthly')

    ratio_matrix_weighted, summary_results_weighted = check_annual_sum(weighted_gas_opcost_cross, weighted_gas_opcost_cross_monthly2)
    weighted_gas_opcost_cross_monthly3 = recalibrate_monthly_production(weighted_gas_opcost_cross_monthly2, monthly_cols, ratio_matrix_weighted)
    plot_cost(weighted_gas_opcost_cross_monthly3, GpuName, 'Scenario', 'monthly', 'Calibrated Smoothed Distributed Weighted Gas OpCost Monthly')

    weighted_gas_opcost_cross_monthly4 = apply_central_moving_average(weighted_gas_opcost_cross_monthly3, monthly_cols, window_size=11)
    plot_cost(weighted_gas_opcost_cross_monthly4, GpuName, 'Scenario', 'monthly', 'Final Smoothed Distributed Weighted Gas OpCost Monthly')


    ratio_matrix_weighted1, summary_results_weighted1 = check_annual_sum(weighted_gas_opcost_cross, weighted_gas_opcost_cross_monthly4)

    weighted_gas_opcost_monthly = pd.melt(weighted_gas_opcost_cross_monthly4, id_vars=['GpuName', 'Scenario'], value_vars=monthly_cols, var_name='Period', value_name='Weighted Gas OpCost')

    print("Unpivoted DataFrame (Weighted Gas OpCost):\n", weighted_gas_opcost_monthly)

    current_time = datetime.datetime.now()
    date_stamp = current_time.strftime("%Y%m%d_%H%M%S")

    file_path = f"C:/Users/NingLin/Documents/output_data_{date_stamp}.xlsx"

    with pd.ExcelWriter(file_path, engine='openpyxl') as writer:
        production_gpu_cross.to_excel(writer, sheet_name='Annual_Cross')
        production_gpu_cross_monthly.to_excel(writer, sheet_name='Monthly_cross')
        production_gpu_cross_monthly2.to_excel(writer, sheet_name='Monthly_cross_smooth')
        production_gpu_cross_monthly3.to_excel(writer, sheet_name='Monthly_cross_smooth_adj')
        production_gpu_cross_monthly4.to_excel(writer, sheet_name='Monthly_cross_smooth_adj2')
        production_gpu_monthly.to_excel(writer, sheet_name='Monthly_table')
        ratio_matrix.to_excel(writer, sheet_name="Ratio")
        summary_results.to_excel(writer, sheet_name="Ratio_Summary")
        ratio_matrix1.to_excel(writer, sheet_name="Ratio1")
        summary_results1.to_excel(writer, sheet_name="Ratio_Summary1")
        weighted_gas_opcost_cross.to_excel(writer, sheet_name="Weighted_Gas_OpCost_Annual_Cross")
        weighted_gas_opcost_cross_monthly.to_excel(writer, sheet_name="Weighted_Gas_OpCost_Monthly_cross")
        weighted_gas_opcost_cross_monthly2.to_excel(writer, sheet_name="Weighted_Gas_OpCost_Monthly_cross_smooth")
        weighted_gas_opcost_cross_monthly3.to_excel(writer, sheet_name="Weighted_Gas_OpCost_Monthly_cross_smooth_adj")
        weighted_gas_opcost_cross_monthly4.to_excel(writer, sheet_name="Weighted_Gas_OpCost_Monthly_cross_smooth_adj2")
        weighted_gas_opcost_monthly.to_excel(writer, sheet_name="Weighted_Gas_OpCost_Monthly_table")
        ratio_matrix_weighted.to_excel(writer, sheet_name="Weighted_Ratio")
        summary_results_weighted.to_excel(writer, sheet_name="Weighted_Ratio_Summary")
        ratio_matrix_weighted1.to_excel(writer, sheet_name="Weighted_Ratio1")
        summary_results_weighted1.to_excel(writer, sheet_name="Weighted_Ratio_Summary1")

    print(f"Data exported successfully to {file_path}")

if __name__ == '__main__':
    main()
  
  

```
